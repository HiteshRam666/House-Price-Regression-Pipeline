```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 
```

```python
train_df = pd.read_csv(r"C:\Users\hites\OneDrive\Desktop\Housing Regression\data\processed\feature_eng_train_df.csv")
eval_df = pd.read_csv(r"C:\Users\hites\OneDrive\Desktop\Housing Regression\data\processed\feature_eng_eval_df.csv")
```

```python
# Target & Features 
target = ["price"] 
X_train = train_df.drop(columns=target)
y_train = train_df[target] 

X_eval = eval_df.drop(columns = target) 
y_eval = eval_df[target]
```

```python
# Standardize 
scaler = StandardScaler() 
X_train_scaled = scaler.fit_transform(X_train) 
X_eval_scaled = scaler.transform(X_eval) 
```

## **Train & Evaluate the models**

### **1. Linear Regression**

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_eval_scaled)
print("Linear Regression")
print(f" MAE: {mean_absolute_error(y_eval, y_pred_lr)}")
print(f" RMSE: {np.sqrt(mean_squared_error(y_eval, y_pred_lr))}")
print(f" R2 Score: {r2_score(y_eval, y_pred_lr)}")
```

```python
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)

y_pred_r = ridge.predict(X_eval_scaled)
print("Ridge Regression")
print(f" MAE: {mean_absolute_error(y_eval, y_pred_r)}")
print(f" RMSE: {np.sqrt(mean_squared_error(y_eval, y_pred_r))}")
print(f" R2 Score: {r2_score(y_eval, y_pred_r)}")
```

```python
lasso = Lasso()
lasso.fit(X_train_scaled, y_train)

y_pred_l = lasso.predict(X_eval_scaled)
print("Lasso Regression")
print(f" MAE: {mean_absolute_error(y_eval, y_pred_l)}")
print(f" RMSE: {np.sqrt(mean_squared_error(y_eval, y_pred_l))}")
print(f" R2 Score: {r2_score(y_eval, y_pred_l)}")
```

```python
elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train_scaled, y_train)
y_pred_elastic = elastic.predict(X_eval_scaled)

print("\nElasticNet Regression:")
print(" MAE:", mean_absolute_error(y_eval, y_pred_elastic))
print(" RMSE:", np.sqrt(mean_squared_error(y_eval, y_pred_elastic)))
print(" R²:", r2_score(y_eval, y_pred_elastic))
```
